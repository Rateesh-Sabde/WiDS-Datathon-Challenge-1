# KFold

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.model_selection import KFold, StratifiedKFold, cross_val_score
from sklearn import linear_model, tree, ensemble

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cnt = 1
for train_index, test_index in kf.split(train_X, train_Y):
    print(f'Fold:{cnt}, Train set: {len(train_index)}, Test set:{len(test_index)}')
    cnt+=1
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC

models=[]

for train_index, val_index in kf.split(train_X, train_Y):
    X_train = train_X.loc[train_index]
    X_val = train_X.loc[val_index]
    Y_train = train_Y.loc[train_index]
    Y_val = train_Y.loc[val_index]
    
    scoring = 'f1'
    svm = SVC(gamma='auto')
    hp = {'C':[0.2, 0.5, 0.8, 1.0, 3.0, 3.5, 4.0, 5.0]}
    clf = GridSearchCV(svm, param_grid = hp, scoring = scoring)
    y = np.array(Y_train)
    #y = y.transpose()
    clf.fit(X_train, y.ravel())
    
    models.append(clf)
    
    print()
    print("-----------------------------------")
    print('Test classification report:')
    y_true, y_pred = Y_val, clf.predict(X_val)
    print(classification_report(y_true, y_pred))

    print('Test AUC-ROC:' + str(roc_auc_score(y_true, y_pred)))

model = np.mean(models, axis = 0)
```

```python
result=[]
for i in range(5):
    result.append(models[i].predict(test_X))
print(result)
```

```python
result = np.array(result)
a = np.mean(result, axis=0)
a = np.where(a>0.5, 1, 0)
result = a
```

```python
ans = pd.DataFrame({'patient_id':np.array(test_ids), 'DiagPeriodL90D': result})
ans.to_csv('submission_3.csv', index = False)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(train_X, train_Y, test_size = 0.2, random_state = 50)
print(X_train.shape)
print(X_val.shape)
print(Y_train.shape)
print(Y_val.shape)
```

```python
result = clf_2.predict(test_X)
ans = pd.DataFrame({'patient_id':np.array(test_ids), 'DiagPeriodL90D': result})
ans.to_csv('submission_4.csv', index = False)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.neural_network import MLPClassifier

models=[]

for train_index, val_index in kf.split(train_X, train_Y):
    X_train = train_X.loc[train_index]
    X_val = train_X.loc[val_index]
    Y_train = train_Y.loc[train_index]
    Y_val = train_Y.loc[val_index]
    
    scoring = 'f1'
    nn = MLPClassifier(solver='adam', alpha=1e-5, random_state=93, max_iter=1000)
    hp = {'hidden_layer_sizes':[1, 2, 3, 4, 5, 6]}
    clf = GridSearchCV(nn, param_grid = hp, scoring = scoring)
    y = np.array(Y_train)
    clf.fit(X_train, y.ravel())
    
    models.append(clf)
    
    print()
    print("-----------------------------------")
    print('Test classification report:')
    y_true, y_pred = Y_val, clf.predict(X_val)
    print(classification_report(y_true, y_pred))

    print('Test AUC-ROC:' + str(roc_auc_score(y_true, y_pred)))
    
result=[]
for i in range(5):
    result.append(models[i].predict(test_X))
print(result)
```
