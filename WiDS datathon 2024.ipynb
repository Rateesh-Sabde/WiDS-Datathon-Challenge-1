```python
train_data = train_data.drop(['patient_id','patient_gender'], axis = 1)
test_ids = test_data['patient_id']
test_data = test_data.drop(['patient_id', 'patient_gender'], axis = 1)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import SimpleImputer

cat_missing_head=[]
for col in train_data.columns.values:
    if(train_data[col].dtypes=="object" and train_data[col].isna().sum()!=0):
        cat_missing_head.append(col)

cat_imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value="unknown")
train_data[cat_missing_head] = cat_imputer.fit_transform(train_data[cat_missing_head])
display(train_data[cat_missing_head])

num_missing_head=[]
for col in train_data.columns.values:
    if(train_data[col].dtypes!="object" and train_data[col].isna().sum()!=0):
        num_missing_head.append(col)
        
temp = train_data[num_missing_head]        
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(temp)
temp = imp.transform(temp)
train_data[num_missing_head] = pd.DataFrame(temp)
display(train_data[num_missing_head])

cat_missing_head=[]
for col in test_data.columns.values:
    if(test_data[col].dtypes=="object" and test_data[col].isna().sum()!=0):
        cat_missing_head.append(col)

cat_imputer = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value="unknown")
test_data[cat_missing_head] = cat_imputer.fit_transform(test_data[cat_missing_head])
display(test_data[cat_missing_head])

num_missing_head=[]
for col in test_data.columns.values:
    if(test_data[col].dtypes!="object" and test_data[col].isna().sum()!=0):
        num_missing_head.append(col)
        
temp = test_data[num_missing_head]        
imp = IterativeImputer(max_iter=15, random_state=0)
imp.fit(temp)
temp = imp.transform(temp)
test_data[num_missing_head] = pd.DataFrame(temp)
display(test_data[num_missing_head])
```
